```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn import svm
from sklearn.model_selection import train_test_split

classes = 4
X,t = make_classification(100, 5, n_informative = 2, n_classes = classes, random_state = 40, n_clusters_per_class = 1)
X_train, X_test, y_train, y_test = train_test_split(X, t, test_size = 0.50, random_state = 40)
model = svm.SVC(C = 1.0, kernel = 'linear', random_state = 40)
model.fit(X_train, y_train)
test = model.predict(X_test)
train = model.predict(X_train)

from sklearn.metrics import accuracy_score
score_test = accuracy_score(test, y_test)
score_train = accuracy_score(train, y_train)

print(score_test)
print(score_train)
```
